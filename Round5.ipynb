```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
from sklearn.model_selection import train_test_split
warnings.filterwarnings('ignore')
```

```python
df = pd.read_csv('/Users/sammy/Desktop/Ironhack/Week2/Day2/lab-customer-analysis-round-3/files_for_lab/csv_files/marketing_customer_analysis.csv')
```

```python
df.head()
```

```python
for column in df:
    if df[column].dtype == np.object:
        x = df[column].unique()
        y = df[column].value_counts()
        fig, ax = plt.subplots(figsize = (12, 9))
        plt.title(column)
        plt.bar(x, y)
        plt.show()    
    elif df[column].dtype == np.number:
        sns.displot(df[column])
```

```python
continuous = ['Customer Lifetime Value']

for i in range(0, len(continuous)):
    plt.figure(i)
    sns.boxplot(x=df[continuous[i]], data=df)
```
